```python
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(client_id=client_id,
                                               client_secret=client_secret,
                                               redirect_uri=redirect_uri,
                                               scope=scope, open_browser=False))
```

```python
def get_random_song_from_file():
    # Path to the song data folder
    song_data_folder = 'musicbrainz'

    # List all files in the song data folder
    all_files = os.listdir(song_data_folder)

    # Filter out only text files (if needed)
    text_files = [file for file in all_files if file.endswith('.txt')]

    # Randomly select a file
    selected_file = random.choice(text_files)

    # Full path to the selected file
    selected_file_path = os.path.join(song_data_folder, selected_file)

    # Read all lines from the selected file
    with open(selected_file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    # Randomly select a line (song name)
    selected_line = random.choice(lines).strip()

    print(selected_line)

    return selected_line
```

```python
def search_song_on_spotify(song_name):
    results = sp.search(q=song_name, type='track', limit=1)
    tracks = results.get('tracks', {}).get('items', [])

    if tracks:
        return tracks[0]  # Return the first search result
    else:
        return None
```

```python
def seek_song():
    song_info = None
    
    while not song_info:
        song_name = get_random_song_from_file()
        print(f"Selected Song: {song_name}")
        
        song_info = search_song_on_spotify(song_name)
        
        if song_info:
            print(f"Found Song on Spotify: {song_info['name']} by {song_info['artists'][0]['name']}")
            print(f"Track URL: {song_info['external_urls']['spotify']}")
        else:
            print("Song not found on Spotify.")

    return song_info
```

```python
features_list = ['danceability', 'energy', 'tempo', 'loudness', 'speechiness', 
                 'acousticness', 'instrumentalness', 'liveness', 'valence']

# Initialize the DataFrame with the specified features and song info columns
df_tracks = pd.DataFrame(columns=['id', 'name', 'artist'] + features_list)

for i in range(100):
    song_info = seek_song()
    song_id = song_info['id']
    audio_features = sp.audio_features([song_id])[0]

    # Collect the necessary information and features
    track_data = {
        'id': song_info['id'],
        'name': song_info['name'],
        'artist': song_info['artists'][0]['name']
    }
    
    for feature in features_list:
        track_data[feature] = audio_features[feature]
    
    track_df = pd.DataFrame([track_data])
    df_tracks = pd.concat([df_tracks, track_df], ignore_index=True)


df_tracks.to_csv('data/testing_data.csv')
print(f"Added 100 songs to the database")
```
